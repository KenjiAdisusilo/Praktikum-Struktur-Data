Bubble Sort

```python
# Bubble Sort
def bubble_sort(arr):
  n = len(arr)
  for i in range(n-1):
    for j in range(n-i-1):
      if arr[j] > arr[j+i]:
        arr[j],arr[j+1] - arr[j+1],arr[j]
  return arr

# Penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = bubble_sort(data.copy()) # Menggunakan copy agar data asli tidak berubah
print("Data sebelum sorting:", data)
print("Data setelah Bubble Sort:", sorted_data)
```

Selection Sort

```python
def insertion_sort(arr):
  n = len(arr)
  for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
      arr[j + 1] = key
  return arr
# Penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Insertion Sort:", sorted_data)
```

Merge Sort

```python
def merge_sort(arr):
  if len(arr) > 1:
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    merge_sort(left_half)
    merge_sort(right_half)
    i = j = k = 0
    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j]:
        arr[k] = left_half[i]
        i += 1
      else:
        arr[k] = right_half[j]
        j += 1
        k += 1
      while i < len(left_half):
        arr[k] = left_half[i]
        i += 1
        k += 1
      while j < len(right_half):
        arr[k] = right_half[j]
        j += 1
        k += 1
  return arr

# Penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Merge Sort:", sorted_data)
```

Quick Sort

```python
def quick_sort(arr):
  if len(arr) <= 1:
    return arr
  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]

  return quick_sort(left) + middle + quick_sort(right)

# Penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Quick Sort:", sorted_data)
```

```python
## Mengurutkan daftar pelanggan

# Quick Sort
def ecom(arr): # fungsi untuk mengurutkan sebuah list menggunakan algoritma quicksort
    if len(arr) <= 1: # jika panjang list kurang dari atau sama dengan 1, kembalikan list itu sendiri karena sudah dianggap terurut
        return arr
    pivot = arr[len(arr) // 2] # memilih elemen tengah sebagai pivot
    left = [x for x in arr if x < pivot] # membuat list left yang berisi elemen yang lebih kecil dari pivot
    middle = [x for x in arr if x == pivot] # membuat list middle yang berisi elemen yang sama dengan pivot
    right = [x for x in arr if x > pivot] # membuat list right yang berisi elemen yang lebih besar dari pivot

    return ecom(left) + middle + ecom(right) # menggabungkan hasil dari left, middle, dan right setelah diurutkan secara rekursif

data = ['Kenji', 'Ayu', 'Budi', 'Dewi', 'Caca','Eka', 'Fajar', 'Gina', 'Hadi', 'Ika'] # data yang akan diurutkan
sorted_data = ecom(data) # memanggil fungsi ecom untuk mengurutkan data dan menyimpan hasilnya dalam variabel sorted_data
print("Data sebelum diurutkan:", data) # mencetak data sebelum diurutkan
print("Data setelah diurutkan:", sorted_data) # mencetak data setelah diurutkan
```
