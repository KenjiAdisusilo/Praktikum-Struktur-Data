```python
## Penyisipan awal
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

# Membuat node
n1 = Node("Asep")
n2 = Node("Udin")
n3 = Node("Toni")

class LinkedList:
    def __init__(self):
        self.headval = None

# Menampilkan Linked List
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval
# Fungsi untuk menambahkan node baru di awal
    def addbegining(self, newdata):
        newnode = Node(newdata)
        newnode.nextval = self.headval
        self.headval = newnode

Li = LinkedList()
Li.headval = n1

# Menghubungkan node pertama dengan node kedua
Li.headval.nextval = n2
# Menghubungkan node ke dua dan ketiga
n2.nextval = n3

print("===Elemen awal===")
Li.listprint()
# Menyisipkan data baru di awal
Li.addbegining("Kenji")
Li.addbegining("Zahra")

print("===Setelah ditambahkan===")
Li.listprint()
```

```python
## Penyisipan Tengah
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

# Membuat Node
n1 = Node("Ken")
n2 = Node("Frey")
n3 = Node("Ipul")

class LinkedList:
    def __init__(self):
        self.headval = None

# Print the Linked List
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval
# Fungsi untuk penyisipan di tengah
    def addinmiddle(self, mid_node, newdata):
        if mid_node is None:
            print("The mentioned node is absent")
            return
        new_node = Node(newdata)
        new_node.nextval = mid_node.nextval
        mid_node.nextval = new_node

Li = LinkedList()
Li.headval = n1

# Menghubungkan node pertama dengan node kedua
Li.headval.nextval = n2
# Menghubungkan node ke dua dan ketiga
n2.nextval = n3

print("===Data awal===")
Li.listprint()
Li.addinmiddle(n1, "Susanto") # (node sebelumnya, data node baru)
Li.addinmiddle(n2, "Supri")
print("===Setelah disisipkan di tengah===")
Li.listprint()
```

```python
## Hapus Node
class Node:
    def __init__(self, dataval = None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

# Menampilkan linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

# Fungsi untuk menambahkan di awal
    def AtBegining(self, newdata):
        NewNode = Node(newdata)
# Mengupdate Linked list setelah di hapus
        NewNode.nextval = self.headval
        self.headval = NewNode

# Fungsi untuk menghapus node dengan kata kunci
    def RemoveNode(self, Removekey):
        HeadVal = self.headval
        if (HeadVal is not None):
            if (HeadVal.dataval == Removekey):
                self.headval = HeadVal.nextval
                HeadVal = None
                return
        while (HeadVal is not None):
            if HeadVal.dataval == Removekey:
                break
            prev = HeadVal
            HeadVal = HeadVal.nextval
        if (HeadVal == None):
            return
        prev.nextval = HeadVal.nextval
        HeadVal = None

# Membuat Linked list dengan nama Li
Li = LinkedList()

# Menambahkan data di node
Li.AtBegining("April")
Li.AtBegining("Maret")
Li.AtBegining("Februari")
print("===Linked list sebelum dihapus===")
Li.listprint()

# Menghapus data dengan kata kunci
Li.RemoveNode("Maret")

print("===Linked list setelah dihapus===")
Li.listprint()
```

```python
## Hapus node akhir

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

# Menampilkan Linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval
# Fungsi untuk menambah node di awal
# Function to add new node in the end of the list
    def AddinEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return
        last = self.headval
        while(last.nextval):
            last = last.nextval
        last.nextval = NewNode

# Fungsi untuk menghapus node di awal
    def RemoveEnd(self):
        last = self.headval
        while(last is not None):
            if last.nextval == None:
                break
            prev = last
            last = last.nextval
        if (last == None):
            return
        prev.nextval = last.nextval
        last = None

# Membuat Linked list dengan nama Li
Li = LinkedList()
# Tambahkan node
Li.AddinEnd("Januari")
Li.AddinEnd("Februari")
Li.AddinEnd("Maret")
print("===Data Awal===")
Li.listprint()

# Hapus node akhir
Li.RemoveEnd()
print("===Data setelah dihapus di akhir===")
Li.listprint()
```

```python
# 1. Doubly Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class DoubleLinkedList:
    def __init__(self):
        self.head = None

    # INSERT DI AKHIR
    def insert_at_end(self, data):
        new_node = Node(data)
        # Jika linked list kosong, node baru menjadi head
        if not self.head:
            self.head = new_node
            return
        # Jika linked list tidak kosong, cari node terakhir
        last = self.head
        while last.next:
            last = last.next
        # Sisipkan node baru di akhir linked list
        last.next = new_node
        new_node.prev = last

    # INSERT SETELAH NODE TERTENTU
    def insert_in_middle(self, prev_node_data, new_data):
        current = self.head
        # Cari node dengan data yang sesuai
        while current and current.data != prev_node_data:
            current = current.next
        # Jika node tidak ditemukan, tampilkan pesan dan keluar
        if current is None:
            print(f"Node '{prev_node_data}' tidak ditemukan.")
            return
        # Jika node ditemukan, buat node baru dan sisipkan setelah node tersebut
        new_node = Node(new_data)
        new_node.next = current.next
        new_node.prev = current
        # Jika node baru bukan node terakhir, perbarui prev dari node berikutnya
        if current.next:
            current.next.prev = new_node
        # Perbarui next dari node saat ini untuk menunjuk ke node baru
        current.next = new_node

    # REMOVE BERDASARKAN KEY
    def removenode(self, removekey):
        if self.head is None:
            print("Linked List kosong.")
            return
        # Cari node dengan data yang sesuai
        current = self.head
        # Cari node dengan data yang sesuai
        while current and current.data != removekey:
            current = current.next
        # Jika node tidak ditemukan, tampilkan pesan dan keluar
        if current is None:
            print(f"Node '{removekey}' tidak ditemukan.")
            return

        # Jika node adalah head
        if current.prev is None:
            self.head = current.next
            if self.head:
                self.head.prev = None
        # Jika node bukan head
        else:
            current.prev.next = current.next
            if current.next:
                current.next.prev = current.prev
        # Indikasi Node berhasil dihapus
        print(f"Node '{removekey}' berhasil dihapus.")

    # DISPLAY MAJU
    def display(self):
        if self.head is None:
            print("Linked List kosong.")
            return
        # Tampilkan semua data dalam linked list
        temp = self.head
        # Tampilkan semua data dalam linked list
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

    # DISPLAY MUNDUR (BUKTI DOUBLY)
    def display_reverse(self):
        # Tampilkan semua data dalam linked list secara terbalik
        if self.head is None:
            print("Linked List kosong.")
            return
        # Cari node terakhir
        temp = self.head
        while temp.next:
            temp = temp.next
        # Tampilkan semua data dalam linked list secara terbalik
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.prev
        print("None")

Li = DoubleLinkedList()
Li.insert_at_end("Kenji")
Li.insert_at_end("Ahmad")
Li.insert_at_end("Somad")

print("=== Awal ===")
Li.display()

print("\n=== Insert Tengah ===")
Li.insert_in_middle("Ahmad", "Budi") # Sisipkan "Budi" setelah "Ahmad"
Li.display()

print("\n=== Hapus Node ===")
Li.removenode("Budi") # Hapus node dengan data "Budi"
Li.display()

print("\n=== Display Reverse ===")
Li.display_reverse()
```

```python
# 2.Circular Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    # INSERT DI AKHIR
    def insert_at_end(self, data):
        new_node = Node(data)

        # Jika list kosong
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return
        # Jika list tidak kosong, cari node terakhir
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        # Sisipkan node baru di akhir circular linked list
        temp.next = new_node
        new_node.next = self.head

    # HAPUS NODE TERAKHIR
    def remove_last(self):
        if self.head is None:
            print("Circular Linked List kosong.")
            return

        # Jika hanya satu node
        if self.head.next == self.head:
            print(f"Node '{self.head.data}' dihapus.")
            self.head = None
            return
        # Jika lebih dari satu node, cari node terakhir dan node sebelumnya
        temp = self.head
        prev = None
        # Cari node terakhir dan node sebelumnya
        while temp.next != self.head:
            prev = temp
            temp = temp.next
        # Hapus node terakhir
        prev.next = self.head
        print(f"Node '{temp.data}' dihapus.")

    # DISPLAY
    def display(self):
        if self.head is None:
            print("Circular Linked List kosong.")
            return
        # Tampilkan semua data dalam circular linked list
        temp = self.head
        # Gunakan loop untuk menampilkan semua node hingga kembali ke head
        while True:
            print(temp.data, end=" -> ") # untuk menampilkan semua data dalam circular linked list
            temp = temp.next
            if temp == self.head:
                break
        print(f"{self.head.data}(kembali ke head)")

CL = CircularLinkedList()

CL.insert_at_end("Kenji")
CL.insert_at_end("Ahmad")
CL.insert_at_end("Somad")

print("=== Circular Linked List Awal ===")
CL.display()

print("\n=== Insert Akhir ===")
CL.insert_at_end("Budi")
CL.display()

print("\n=== Hapus Akhir ===")
CL.remove_last()
CL.display()
```
