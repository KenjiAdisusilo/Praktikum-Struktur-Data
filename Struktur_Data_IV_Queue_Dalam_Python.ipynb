```python
## Queue dengan List ##
queue = []
# Demonstrasi queue dengan list
queue.append('k')
queue.append('e')
queue.append('n')
queue.append('j')
queue.append('i')

print('===Queue===')
print(queue)

# pop(0) fungsi untuk dequeue(menghapus) elemen dari queue dalam konsep FIFO
print("===ELemen yang dequeued dari queue===")
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('===Queue setelah dihapus===')
print(queue)

# Jika kosong dan memberikan peritah queue.pop(0) maka akan terjadi eror
```

```python
## Queue dengan collection.deque ##
from collections import deque
queue = deque()

# append() fungsi untuk enqueue(menambah) elemen di queue
queue.append('A')
queue.append('S')
queue.append('A')
queue.append('D')
queue.append('E')
queue.append('L')

print('===Queue===')
print(queue)

# popleft() fungsi untuk dequeue(menghapus) elemen dari stack dengan konsep LIFO
print('===Elemen yang Dihapus===')
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print('===Queue setelah Dihapus===')
print(queue)

# Akan eror jika stack kosong, tetapi mencoba menghapus
```

```python
## Queue dengan queue.Queue ##
from queue import Queue

# Inisiasi queue
qu = Queue(maxsize=5)



# put() fungsi untuk menambah elemen
qu.put('A')
qu.put('s')
qu.put('e')
qu.put('p')

print('===Ukuran Queue===')
print('Full =', qu.full()) # qu.full() menampilkan apakah full atau tidak
print('Size =',qu.qsize()) # qsize() menunjukan jumlah elemen dalam queue

# get() fungsi untuk mengambil dan menghapus
print('===Get Queue===')
print(qu.get())
print(qu.get())
print(qu.get())
print(qu.get())

print('===Ukuran Queue Akhir===')
print('Full =', qu.full())
print('Size =',qu.qsize())
```

```python
## Queue dengan Single Linked List ##
class Queue:
  def __init__(self):
    self.queue = list()

  def add(self,data):
    if data not in self.queue:
      self.queue.insert(0,data)
      return data
    return False

  def remove(self):
    if len(self.queue)>0:
      return self.queue.pop()
    return ("Queue Kosong!")

  def size(self):
    return len(self.queue)

# Deklarasi kan linked list sbg Qu
Qu= Queue()
print("==Elemen yang Ditambahkan==")
print(Qu.add("Ken"))
print(Qu.add("Cuk"))
print(Qu.add("Lek"))
print(Qu.add("Bro"))
print("Size=",Qu.size())
print("==Elemen yang dihapus==")
print(Qu.remove())
print(Qu.remove())
print("Size=",Qu.size())
```
