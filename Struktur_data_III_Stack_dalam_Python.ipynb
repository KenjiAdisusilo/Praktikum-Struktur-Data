```python
# demonstrasi stack menggunakan list
stack = [] # membuat stack kosong

# menambahkan elemen ke stack dengan append()
stack.append("Ken")
stack.append("Zee")
stack.append("Jono")
stack.append("Udin")
stack.append("Yoko")

print("===Stack dengan List===")
print(stack)

# menghapus elemen terakhir dari stack dengan pop()
print("===Pop Elemen dari Stack===")
print(stack.pop())
print(stack.pop())
print(stack.pop())

print("===Stack Setelah Pop===")
print(stack)
```

```python
# Demonstrasi implementasi stack dengan collection.deque
from collections import deque

stack = deque()  # Membuat stack menggunakan deque

# Menambahkan elemen ke stack dengan append()
stack.append(1)
stack.append(2)
stack.append(3)
print("===Stack setelah menambahkan elemen===")
print(stack)

# Menghapus elemen dari stack dengan pop()
print("===Elemen yang dihapus===")
print(stack.pop())
print(stack.pop())

print("===Stack setelah menghapus elemen===")
print(stack)

# Akan terjadi eror saat coba menampilkan stack yang kosong
```

```python
# Demonstrasi implementasi stack dengan queue
from queue import LifoQueue
# inisialisasi stack
stack = LifoQueue(maxsize=5)

print("===Stack Awal===")
# qsize() untuk menampilkan jumlah elemen dalam stack
print(f"Jumlah elemen dalam stack: {stack.qsize()}") # Output: 0

# put() untuk menambahkan elemen ke dalam stack
stack.put('k')
stack.put('e')
stack.put('n')
stack.put('j')
stack.put('i')

print("\n===Setelah menambahkan elemen ke dalam stack===")
print(f"Elemen di stack: {stack.queue}") # Output: ['k', 'e', 'n', 'j', 'i']
print(f"Full: {stack.full()}") # menunujukkan apakah stack penuh atau tidak, Output: True
print(f"Jumlah elemen dalam stack: {stack.qsize()}") # Output: 3

# get() untuk menghapus elemen dari stack
print("\n===Menghapus elemen dari stack===")
print(f"Elemen yang dihapus:")
print(stack.get()) # Output: 'i'
print(stack.get()) # Output: 'j'

print("\n===Setelah menghapus elemen dari stack===")
print(f"Elemen di stack: {stack.queue}") # Output: ['k', 'e', 'n']
print(f"Empty: {stack.empty()}") # menunujukkan apakah stack kosong atau tidak, Output: False
print(f"Jumlah elemen dalam stack: {stack.qsize()}") # Output: 3




```

```python
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class stack:
  def __init__(self):
    self.head = Node(None)
    self.size = 0

  # string representasi dari stack
  def __str__(self):
    cur = self.head.next
    out = ""
    while cur:
      out += str(cur.value) + "->"
      cur = cur.next
    return out[:-2]

  # Fungsi untuk menampilkan ukuran stack
  def getSize(self):
    return self.size

  # Fungsi untuk memeriksa apakah stack kosong
  def isEmpty(self):
    return self.size == 0

  # Fungsi untuk menampilkan elemen teratas/peek
  def peek(self):
    # Jika peek stack yang kosong, lakukan pengecuaian
    if self.isEmpty():
      raise Exception("Stack Kosong....")
      return self.head.next.value # Mengembalikan nilai dari elemen teratas stack

  # Fungsi untuk menambahkan elemen ke stack
  def push(self, value):
    node = Node(value) # Membuat node baru dengan nilai yang diberikan
    node.next = self.head.next # Menyambungkan node baru ke elemen teratas stack saat ini
    self.head.next = node # Menjadikan node baru sebagai elemen teratas stack
    self.size += 1 # Menambah ukuran stack

  # Fungsi untuk menghapus elemen teratas stack
  def pop(self):
    # Jika pop stack yang kosong, lakukan pengecuaian
    if self.isEmpty():
      raise Exception("Stack Kosong....")
    remove = self.head.next # Menyimpan elemen teratas stack yang akan dihapus
    self.head.next = self.head.next.next # Menyambungkan head ke elemen berikutnya setelah elemen teratas
    self.size -= 1 # Mengurangi ukuran stack
    return remove.value # Mengembalikan nilai dari elemen yang dihapus

# Driver code untuk menguji implementasi stack
if __name__ == "__main__":
  s = stack()
  for i in range(1, 11):
    s.push(i) # Menambahkan elemen 1 hingga 10 ke stack
  print("Stack:", s) # Menampilkan isi stack

  for _ in range(5):
    print("Pop:", s.pop()) # Menghapus dan menampilkan elemen teratas stack
  print("Stack setelah pop:", s) # Menampilkan isi stack setelah beberapa elemen dihapus
  print(f"Stack:{s}, Size: {s.getSize()}") # Menampilkan isi stack dan ukuran saat ini



```
