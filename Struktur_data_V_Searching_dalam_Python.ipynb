```python
# Linear Search
def linear_search(arr,target):
  for i in range(len(arr)):
    if arr[i] == target:
      return i
  return -1

arr = [12,45,30,21,14,22,28,33]
target = int(input("Masukan nilai yang ingin dicari:..."))

result = linear_search(arr,target)

if result != -1:
  print(f"Elemen ditemukan pada indeks {result}")
else:
  print("Elemen tidak ditemukan")
```

```python
# Binary Search
def binary_search(arr, target):
  low = 0
  high = len(arr) - 1
  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
  return -1

arr = [1990, 1992, 1995, 1997, 2005, 2009, 2011, 2012]
target = int(input("Masukkan tahun lahir Anda (YYYY): "))
result = binary_search(arr, target)
if result != -1:
  print(f"Elemen ditemukan pada indeks {result}")
else:
  print("Elemen tidak ditemukan")
```

```python
import math

def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    while arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1
    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i
    return -1

# Data harus urut (alfabet)
arr = ["Aldi", "Nabila", "Kenji", "Tono", "Jihan", "Naura", "Jefri", "Asep"]
target = input("Masukkan nama Anda: ")
result = jump_search(arr, target)

if result != -1:
    print(f"Elemen ditemukan pada indeks {result}")
else:
    print("Elemen tidak ditemukan")
```

```python
import random
import math

# 1. Membuat Data
data = random.sample(range(1, 1001), 100)
data.sort()  # harus urut untuk binary & jump search

# ambil target (elemen tengah)
target = data[len(data)//2]

print("Data:", data)
print("Target yang dicari:", target)
print()

# 2. Linear Search
def linear_search(arr, target):
    perbandingan = 0

    for i in range(len(arr)):
        perbandingan += 1
        if arr[i] == target:
            return i, perbandingan

    return -1, perbandingan

# 3. Binary Search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    perbandingan = 0

    while low <= high:
        mid = (low + high) // 2
        perbandingan += 1

        if arr[mid] == target:
            return mid, perbandingan
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, perbandingan

# 4. Jump Search
def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    perbandingan = 0

    # lompat per blok
    while arr[min(step, n) - 1] < target:
        perbandingan += 1
        prev = step
        step += int(math.sqrt(n))

        if prev >= n:
            return -1, perbandingan

    # cek dalam blok
    for i in range(prev, min(step, n)):
        perbandingan += 1
        if arr[i] == target:
            return i, perbandingan

    return -1, perbandingan

# 5. Menjalankan Pencarian
lin_index, lin_count = linear_search(data, target)
bin_index, bin_count = binary_search(data, target)
jump_index, jump_count = jump_search(data, target)

# 6. Output Hasil
print("=== HASIL PENCARIAN ===")
print("Linear Search")
print("Index:", lin_index)
print("Jumlah perbandingan:", lin_count)
print()

print("Binary Search")
print("Index:", bin_index)
print("Jumlah perbandingan:", bin_count)
print()

print("Jump Search")
print("Index:", jump_index)
print("Jumlah perbandingan:", jump_count)
```

```python
## STUDI KASUS 1 ##
# Program untuk mencari nilai minimum dan maksimum dalam array

class Pair:
    def __init__(self):
        self.min = 0
        self.max = 0

def getMinMax(arr, n):
    minmax = Pair()

    # Jika hanya ada 1 elemen
    if n == 1:
        minmax.max = arr[0]
        minmax.min = arr[0]
        return minmax

    # Inisialisasi awal
    if arr[0] > arr[1]:
        minmax.max = arr[0]
        minmax.min = arr[1]
    else:
        minmax.max = arr[1]
        minmax.min = arr[0]

    # Loop dari elemen ke-3
    for i in range(2, n):
        if arr[i] > minmax.max:
            minmax.max = arr[i]
        elif arr[i] < minmax.min:
            minmax.min = arr[i]

    return minmax

# Driver Code
if __name__ == "__main__":
    arr = [1000, 11, 445, 1, 330, 3000]
    arr_size = len(arr)

    minmax = getMinMax(arr, arr_size)

    print("Minimum element is", minmax.min)
    print("Maximum element is", minmax.max)
```

```python
## STUDI KASUS 2 ##
def printTwoElements(arr):
    n = len(arr)

    # Membuat array bantu (temp) berisi 0
    temp = [0] * n

    repeatingNumber = -1
    missingNumber = -1

    # Mencari angka yang duplikat
    for i in range(n):
        temp[arr[i] - 1] += 1
        if temp[arr[i] - 1] > 1:
            repeatingNumber = arr[i]

    # Mencari angka yang hilang
    for i in range(n):
        if temp[i] == 0:
            missingNumber = i + 1
            break

    print("The repeating number is", repeatingNumber)
    print("The missing number is", missingNumber)


# Contoh penggunaan
arr = [7, 3, 4, 5, 5, 6, 2]
printTwoElements(arr)
```

```python
## STUDI KASUS 3 ##
def ada_pasangan(arr, X):
    n = len(arr)

    # Cari titik rotasi (pivot)
    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            break

    # Pointer kiri = elemen terkecil
    l = (i + 1) % n
    # Pointer kanan = elemen terbesar
    r = i

    # Metode two pointer
    while l != r:
        if arr[l] + arr[r] == X:
            return True
        elif arr[l] + arr[r] < X:
            l = (l + 1) % n
        else:
            r = (r - 1 + n) % n

    return False


# Contoh pengujian
print(ada_pasangan([11, 15, 6, 8, 9, 10], 16))  # True
print(ada_pasangan([11, 15, 26, 38, 9, 10], 35))  # True
print(ada_pasangan([11, 15, 26, 38, 9, 10], 45))  # False
```
